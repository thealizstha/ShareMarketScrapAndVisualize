```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://merolagani.com/Floorsheet.aspx#0','html.parser')
soup = requests.get(url)
#soup.text
raw_code = BeautifulSoup(soup.text,'lxml')
raw_code
```

```python
table_code = raw_code.table
table_code
```

```python
tr_tags = table_code.find_all('tr')
tr_tags
```

```python
l = []
final = []
for tag in tr_tags:
   x = tag.text.split('\n') # for removing white space in starting and ending we can use strip() also.
   if x:
   #x = tag.text
      l.append(x)
l = [[item.strip() for item in sublist if item.strip()] for sublist in l] #removing white spaces    
print(l)

```

```python
import csv
file = open('share_floor_data.csv','w')
y = csv.writer(file)
y.writerows(l)
file.close()
```
